```python
import json
import os

# --- DEBUGGING: Where am I? ---
print(f" Current Working Directory: {os.getcwd()}")

# --- STEP 1: Define the Absolute Path ---
# In Colab, the main folder is always /content/
default_path = "/content/config.json"

def create_sample_json(filename):
    sample_data = {
        "server_name": "colab-server-01",
        "database": {"host": "localhost", "port": 5432},
        "features": {"debug_mode": True},
        "thresholds": {"cpu_limit": 85.5}
    }
    try:
        with open(filename, "w") as f:
            json.dump(sample_data, f, indent=4)
        print(f" Created sample file at: '{filename}'")
    except Exception as e:
        print(f" Failed to create file: {e}")

def load_settings(file_path):
    print(f"\n Loading from: {file_path}")
    try:
        with open(file_path, "r") as f:
            data = json.load(f)
            print(" Success! File loaded.")
            return data
    except FileNotFoundError:
        print(f"Error: File not found at {file_path}")
        return None

# --- MAIN ---
# 1. Create the file in the specific Colab folder
if not os.path.exists(default_path):
    create_sample_json(default_path)

# 2. Ask user (or just press Enter to use the safe path)
user_input = input(f"Enter path (default: {default_path}): ").strip()
target_path = user_input if user_input else default_path

# 3. Load
settings = load_settings(target_path)

if settings:
    print(f"Server Name: {settings['server_name']}")
```

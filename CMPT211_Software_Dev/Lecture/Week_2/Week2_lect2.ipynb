Part 2: Python Crash Course (The "Cleaning" Logic)

Before we create files on your computer, let's learn the Python commands right here.

The Scenario: You have a list of user data, but it is messy. There are spaces where there shouldn't be.

    Bad Data: " Alice "

    Good Data: "Alice"

Run the code cell below to generate our "Dummy Data".

```python
# [INSTRUCTOR NOTE: Run this to set up the student's workspace]

# 1. This is a Python "List". It holds multiple pieces of text (Strings).
# Notice the extra spaces inside the quotes!
raw_strings = [
    "   admin_user ",
    "  guest_2026",
    "root_access  ",
    "   system_log "
]

print("Load Data:")
print(raw_strings)
```

The Solution: Loops and Strip

To clean this data, we need two tools:

    The Loop (for): This tells Python to "pick up each item one by one."

    The Method (.strip()): This is a built-in tool that cuts off whitespace from the start and end of a word.

Activity: Look at the code below. This is the solution to your first task.

    We look at user.

    We apply .strip().

    We print the result.

Run the cell below to see it work.


```python
# [INSTRUCTOR DEMO / ANSWER KEY]

print("✅ CLEANING IN PROGRESS...\n")

# This loop reads every item in our list 'raw_strings'
# We temporarily call the current item 'user'
for user in raw_strings:

    # 1. Apply the strip() tool to remove spaces
    clean_user = user.strip()

    # 2. Print the before and after
    print(f"Original: '{user}'  --->  Cleaned: '{clean_user}'")

print("\nSuccess! Logic verified.")
```
